```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.

df1 = pd.read_csv("Resources/athletic_sales_2020.csv")
df2 = pd.read_csv("Resources/athletic_sales_2021.csv")
```

```python
# Display the 2020 sales DataFrame
df1.head()
```

```python
# Display the 2021 sales DataFrame
df2.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
df1.info()
```

```python
# Check the 2021 sales data types.
df2.info()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
df_combined = pd.concat([df1, df2], ignore_index=True)
df_combined.head()
```

```python
# Check if any values are null.
df_combined.isnull().sum()
```

```python
# Check the data type of each column
df_combined.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
df_combined["invoice_date"] = pd.to_datetime(df_combined["invoice_date"])
df_combined.head()
```

```python
# Confirm that the "invoice_date" data type has been changed.
df_combined.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
df_products_sold = df_combined.groupby(["region", "state", "city"]).agg({"units_sold": "sum"}).reset_index()
df_products_sold = df_products_sold.rename(columns={"units_sold": "Total_Products_Sold"})
df_products_sold = df_products_sold.sort_values(by="Total_Products_Sold", ascending=False)
# Show the top 5 results.
df_products_sold.head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
df_products_sold_pivottable = pd.pivot_table(df_combined, values='units_sold', index=['region', 'state', 'city'], aggfunc='sum').reset_index()
# Rename the "units_sold" column to "Total_Products_Sold"
df_products_sold_pivottable = df_products_sold_pivottable.rename(columns={"units_sold": "Total_Products_Sold"})
df_products_sold_pivottable = df_products_sold_pivottable.sort_values(by="Total_Products_Sold", ascending=False)
# Show the top 5 results.
df_products_sold_pivottable.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
df_total_sales = df_combined.groupby(["region", "state", "city"]).agg({"total_sales": "sum"}).reset_index()
df_total_sales = df_total_sales.rename(columns={"total_sales": "Total Sales"})
df_total_sales = df_total_sales.sort_values(by="Total Sales", ascending=False)

# Show the top 5 results.
df_total_sales.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
df_total_sales_pivottable = pd.pivot_table(df_combined, values='total_sales', index=['region', 'state', 'city'], aggfunc='sum').reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
df_total_sales_pivottable = df_total_sales_pivottable.rename(columns={"total_sales": "Total_Sales"})
df_total_sales_pivottable = df_total_sales_pivottable.sort_values(by="Total_Sales", ascending=False)

# Show the top 5 results.
df_total_sales_pivottable.head()
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
df_total_sales_retailer = df_combined.groupby(["region", "state", "city", "retailer"]).agg({"total_sales": "sum"}).reset_index()
df_total_sales_retailer = df_total_sales_retailer.rename(columns={"total_sales": "Total Sales"})
df_total_sales_retailer = df_total_sales_retailer.sort_values(by="Total Sales", ascending=False)

# Show the top 5 results.
df_total_sales_retailer.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
df_total_sales_retailer_pt = pd.pivot_table(df_combined, values='total_sales', index=['region', 'state', 'city', 'retailer'], aggfunc='sum').reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
df_total_sales_retailer_pt = df_total_sales_retailer_pt.rename(columns={"total_sales": "Total_Sales"})
df_total_sales_retailer_pt = df_total_sales_retailer_pt.sort_values(by="Total_Sales", ascending=False)

# Show the top 5 results.
df_total_sales_retailer_pt.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
print(df_combined.columns)
```

```python
# Filter the sales data to get the women's athletic footwear sales data.
df_womens_footwear = df_combined[df_combined["product"] == "Women's Athletic Footwear"]
df_womens_footwear.head()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
df_womens_footwear_sold = df_womens_footwear.groupby(["region", "state", "city", "retailer"]).agg({"units_sold": "sum"}).reset_index()
df_womens_footwear_sold = df_womens_footwear_sold.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"})
df_womens_footwear_sold = df_womens_footwear_sold.sort_values(by="Womens_Footwear_Units_Sold", ascending=False)

# Show the top 5 results.
df_womens_footwear_sold.head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
df_womens_footwear_sold_pt = pd.pivot_table(df_womens_footwear, values='units_sold', index=['region', 'state', 'city', 'retailer'], aggfunc='sum').reset_index()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
df_womens_footwear_sold_pt = df_womens_footwear_sold_pt.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"})
df_womens_footwear_sold_pt = df_womens_footwear_sold_pt.sort_values(by="Womens_Footwear_Units_Sold", ascending=False)

# Show the top 5 results.
df_womens_footwear_sold_pt.head()
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
df_sales_by_date = pd.pivot_table(df_womens_footwear, values='total_sales', index=['invoice_date'], aggfunc='sum').reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
df_sales_by_date = df_sales_by_date.rename(columns={"total_sales": "Total Sales"})

# Show the table.
df_sales_by_date.head(10)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
df_sales_by_date_resampled = df_sales_by_date.set_index('invoice_date').resample('D').sum().reset_index()

# Sort the resampled pivot table in descending order on "Total Sales".
df_sales_by_date_resampled = df_sales_by_date_resampled.sort_values(by="Total Sales", ascending=False)
df_sales_by_date_resampled.head(10)
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
df_sales_by_date_resampled_weekly = df_sales_by_date.set_index('invoice_date').resample('W').sum().reset_index()

# Sort the resampled pivot table in descending order on "Total Sales".
df_sales_by_date_resampled_weekly = df_sales_by_date_resampled_weekly.sort_values(by="Total Sales", ascending=False)
df_sales_by_date_resampled_weekly.head(10)
```
